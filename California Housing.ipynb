```python
from sklearn.model_selection import train_test_split
```

```python
# using the train test split function
x_train, x_test, y_train, y_test = train_test_split(df.drop(columns=['median_house_value']),df['median_house_value'], random_state=12, test_size=0.2, shuffle=True)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
regr = LinearRegression()
```

```python
regr.fit(x_train, y_train)
print(regr.score(x_test, y_test))
```

```python
y_pred = regr.predict(x_test)
```

```python
regr.score(x_train, y_train)
```

```python
from sklearn.metrics import r2_score
```

```python
r2 = r2_score(y_test, y_pred)
```

```python
print(r2)
```

```python
from sklearn.metrics import mean_absolute_percentage_error
```

```python
mape = mean_absolute_percentage_error(y_test, y_pred)
```

```python
print(mape)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, y_pred, edgecolors=(0, 0, 0), alpha=0.2)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "k-", lw=2)
ax.set_xlabel("Actual")
ax.set_ylabel("Predicted")
ax.set_title("Model evaluation")
plt.show()
```

### Hyper Tuned Catboost 

```python
model_CBR_hyper = ctb.CatBoostRegressor()
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid_cbr_hyper = GridSearchCV(estimator=model_CBR_hyper, param_grid = parameters, cv = 2, n_jobs=-1)
grid_cbr_hyper.fit(x_train, y_train)
```

### Hyper Tuned XGBoost

```python
xgb1 = xgb.XGBRegressor()
```

```python
parameters = {'nthread':[4], #when use hyperthread, xgboost may become slower
              'objective':['reg:squarederror'],
              'learning_rate': [.03, 0.05, .07], #so called `eta` value
              'max_depth': [5, 6, 7],
              'min_child_weight': [4],
              'silent': [1],
              'subsample': [0.7],
              'colsample_bytree': [0.7],
              'n_estimators': [500]}
```

```python
xgb_grid = GridSearchCV(xgb1,
                        parameters,
                        cv = 2,
                        n_jobs = 5,
                        verbose=True)
```

```python
xgb_grid.fit(x_train,
         y_train)
```

```python
print(xgb_grid.best_score_)
print(xgb_grid.best_params_)
```

```python
preds_hyper_xgb = xgb_grid.predict(x_test)
```

```python
r2_xgb_hyper = r2_score(y_test, preds_hyper_xgb)
```

```python
print(r2_xgb_hyper)
```

```python
mape_xgb_hyper = mean_absolute_percentage_error(y_test, preds_hyper_xgb)
```

```python
print(mape_xgb_hyper)
```

```python
fig, ax = plt.subplots()
ax.scatter(y_test, preds_hyper_xgb, edgecolors=(0, 0, 0), alpha=0.2)
ax.plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], "k-", lw=2)
ax.set_xlabel("Actual")
ax.set_ylabel("Predicted")
ax.set_title("Model evaluation")
plt.show()
```
